lower() 將字母轉換為小寫
axis = 1 維度為1

title_overviews(:5)取前5個值

re.sub()取代
r'[^a-zA-Z ]+',' ' 不是a-z或A-Z 或空白 並以空白取代 

```python
import re
def get_title_overview(df):
    #
    title_overviews=df.apply(lambda x:x['title'].lower()+' '+x['overview'].lower(),axis=1).values
    
    title_overviews=[re.sub(r'[^a-zA-Z]+',' ',title_overviews) for title_overviews in title_overviews]
    
    title_overviews=[title_overviews.split(' ') for title_overviews in title_overviews ]
    
    return title_overviews
title_overviews=get_title_overview(df)
title_overviews[:5]
```

```python
max_words=10000
from collections import Counter
counter=Counter(np.hstack(title_overviews))
```

-x[1]原本是由小到大 加了負號變成由大到小(出現頻率)

```python
word2index = {unique_word: idx for idx, (unique_word, count ) in enumerate(sorted(counter.items(),key=lambda x: -x[1] ))if idx<10000}
word2index
```

len維度 第一維是title_overviews電影名稱
for word in title_overview 印出每個字

```python
count_matrix = np.zeros([len(title_overviews),max_words], dtype=np.int32)

for idx, title_overview in enumerate(title_overviews):
    for word in title_overview:
        if word in word2index:
            count_matrix[idx][word2index[word]] += 1      
        
count_matrix    
```

把做好的count_matrix丟進TFIDF中

```python
transformer = TfidfTransformer()
tfidf_matrix = transformer.fit_transform(count_matrix)
tfidf_matrix.shape
```

```python
tfidf_matrix = tfidf_matrix.toarray()
```

```python
(tfidf_matrix ** 2).sum(axis  =1)
```

推薦步驟
num_movies 電影數
max_words 字詞數

np.argsort 由小排到大
np.flip 由大排到小

most_similar_items = most_similar_items [most_similar_items != idx]
=>推薦的不能=自己
[:top_n]取前n個

```python
def get_most_similar_items(tfidf_matrix, idx, top_n):
    
    #1d array (max_words,)
    tfidf_matrix[idx]
    
    #2d array (num_movies,max_words)
    tfidf_matrix
    
    #(num_movies, max_words) * (max_words, 1)->(num_movies, 1)->(num_movies)
    scores = np.matmul(tfidf_matrix,tfidf_matrix[idx].reshape(-1, 1)).reshape(-1)
    
    most_similar_items = np.flip(np.argsort(scores))
    
    most_similar_items = most_similar_items [most_similar_items != idx][:top_n]
    
    return most_similar_items
```
